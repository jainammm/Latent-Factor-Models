```python
import pandas as pd 
import numpy as np
```

```python
dataset = pd.read_csv('ratings.csv')
dataset.head()
```

```python
movieTitles = pd.read_csv('movies.csv')
movieTitles.head()
```

```python
dataset = pd.merge(dataset, movieTitles, on='movieId')
dataset.head()
```

```python
movieMatrix = dataset.pivot_table(index='userId', columns='movieId', values='rating')
movieMatrix.head()
```

```python
R = movieMatrix.transpose().fillna(movieMatrix.transpose().mean())
R = R.transpose()
R.head()
```

```python
R = np.array(R)
R.shape
```

```python
numUsers, numItems = R.shape
```

```python
np.linalg.matrix_rank(R)
```

```python
K = 400
```

```python
alpha = 0.000001
```

```python
U = np.random.randn(numUsers, K)
U
```

```python
V = np.random.randn(numItems, K)
V
```

```python
where_are_NaNs = np.isnan(movieMatrix)
where_are_NaNs = np.array(where_are_NaNs)
where_are_NaNs
```

```python
samples = np.argwhere(where_are_NaNs == False)
samples
```

```python
for it in range(100000):
    np.random.shuffle(samples)
        
    # SGD
    for i, j in samples:
        prediction = U[i, :].dot(V[j, :].transpose())
        e = R[i,j] - prediction
        
        # Update
        U[i, :] += alpha * (e * V[j, :])
        V[j, :] += alpha * (e * U[i, :])
    
    error = R - np.matmul(U, V.transpose())
    error[where_are_NaNs] = 0
    print("Iteration: ", it, " ---> Frobenius norm: ", np.linalg.norm(error))
```
